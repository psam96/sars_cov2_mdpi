```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

data = pd.read_csv("./data/monthly_avg_cases_deaths.csv")

fig, ax = plt.subplots(figsize = (9, 5.5))
ax.plot(data.date, data.avg_cases,  label = "Average New Cases", color = "Green")
ax.margins(x = 0)

ax.set_ylabel("Number of new cases")
ax.set_xlabel("Time")
plt.xticks(rotation = 90)

ax2 = ax.twinx()
ax2.plot(data.date, data.avg_deaths, label = "Average New Deaths", color = "Red")
ax2.set_ylabel("Number of deaths", rotation = -90, labelpad = 15)
ax2.set_yticks(np.arange(0,200,20))
ax2.margins(x = 0)

ax2.axvline(x = "Jan 2021", label = "Delta Variant Introduced", color = "purple", ls = "-.")

fig.legend(bbox_to_anchor = (0.345, 0.965))
ax.grid()

plt.tight_layout()
plt.savefig("graph_avg_death_cases_added_details.png", dpi = 600)
plt.show()
```
