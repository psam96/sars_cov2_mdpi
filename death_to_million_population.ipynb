```python
from matplotlib import pyplot as plt
import numpy as np

country = [["Bosnia and H.", "Italy" , "Spain", "Norway", "Iceland"],
          ["USA", "Canada", "Jamaica", "Dominica"], 
          ["Georgia", "Iran", "Sri Lanka", "India", "Bangladesh", "Bhutan"], 
          ["Peru", "Brazil", "Venezuela"],
          ["Tunisia", "South Africa", "Congo"]]

ratio = [[3159, 2160, 1836, 154, 96],
        [2084, 719, 603, 138],
        [2166, 1378, 568, 319, 163, 4],
        [5936, 2757, 152],
        [2049, 1432, 32]]

fig, ax = plt.subplots(figsize = (11, 8))

ax.bar(country[0], ratio[0], color = "#b722c7", width = 0.9, label = "Europe")
ax.bar(country[1], ratio[1], color = "#71c7ad", width = 0.9, label = "North America")
ax.bar(country[2], ratio[2], color = "#e0437d", width = 0.9, label = "Asia")
ax.bar(country[3], ratio[3], color = "#c99642", width = 0.9, label = "South America")
ax.bar(country[4], ratio[4], color = "#a0c722", width = 0.9, label = "Africa")

ax.set_xlabel("Countries")
ax.set_ylabel("Deaths per million population")
ax.margins(x = 0.005)
ax.set_yticks(np.arange(0, 6500, 500))

ax.axhline(y = 371, color = "black", ls = "-.", label = "Deaths per million population in Nepal")
ax.legend()

plt.xticks(rotation = 90)

plt. tight_layout()
plt.savefig("death_per_million_population.png", dpi = 600)
plt.show()
```
