```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

data = pd.read_csv("./data/monthly_avg_cases_deaths.csv")

fig, ax = plt.subplots(figsize = (9,4))
ax.plot(data.date, data.death_to_case, color = "red", label = "Death to Cases Ratio")
ax.set_xlabel("Time")
ax.set_ylabel("Ratio")
ax.set_yticks(np.arange(0,0.3, 0.05))
ax.axvline(x = "Jan 2021", label = "Delta Variant Introduced", color = "purple", ls = "-.")
ax.margins(x = 0)

plt.grid(True)
plt.xticks(rotation = 90)
plt.legend(loc = "upper left")

plt.tight_layout()
plt.savefig("death_to_case_ratio.png", dpi = 600)
plt.show()
```
