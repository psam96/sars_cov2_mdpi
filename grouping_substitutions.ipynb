```python
from matplotlib import pyplot as plt
import numpy as np

data = [["S:A570S", "S:D936H", "S:K1073N", "S:S477N"],
       ["S:A570D", "S:D1118H", "S:N501Y", "S:S982A", "S:T716I"],
       ["S:T478K", "S:T19R", "S:T95I", "S:D950N", "S:A222V", "S:W258L", "S:K417N", "S:R158G", "S:V1104L"],
       ["S:P681H"],
       ["S:L452R", "S:P681R", "S:G142D", "S:E484Q"],
       ["S:D614G"]]

value = [[3, 3, 3, 3], 
        [11, 11, 11, 11, 11],
        [101, 100, 26, 25, 23, 12, 8, 6, 5],
        [14],
        [103, 103, 10, 3],
        [125]]
        
fig, ax = plt.subplots(figsize = (11,6))

ax.bar(data [0], value [0], color = "#1b3282", width = 0.9, label = "20A")
ax.bar(data [1], value [1], color = "#71c7ad", width = 0.9, label = "Alpha")
ax.bar(data [2], value [2], color = "#e0437d", width = 0.9, label = "Delta")
ax.bar(data [3], value [3], color = "#c99642", width = 0.9, label = "20A and Alpha")
ax.bar(data [4], value [4], color = "#b722c7", width = 0.9, label = "Delta and Kappa")
ax.bar(data [5], value [5], color = "#a0c722", width = 0.9, label = "20A, 20B, Alpha, Delta and Kappa ")


ax.set_xlabel("Amino acid substitutions")
ax.set_ylabel("Number of sequences")
ax.set_yticks(np.arange(0,140,10))

plt.xticks(rotation = 90)
plt.margins(x = 0.005)
plt.legend()

plt.tight_layout()
plt.savefig("s_protein_substitution.png", dpi = 500)
plt.show()
```
