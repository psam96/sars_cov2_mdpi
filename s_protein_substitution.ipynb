```python
#Program to prepare the s protein substitution vs number of sequences graph

from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i]//0.98,y[i], ha = "center")

data = pd.read_csv("./data/s_protein_substitution.csv")

x = data.substitution
y = data.number

fig, ax = plt.subplots(figsize = (14,8))
addlabels(x, y)

ax.bar(x, y, color = "green")
ax.set_xlabel("Amino acid substitutions")
ax.set_ylabel("Number of sequences")
plt.xticks(rotation = 90)
plt.yticks(np.arange(0,140,10))
ax.margins(x = 0.005)

plt.tight_layout()
plt.show()
```
