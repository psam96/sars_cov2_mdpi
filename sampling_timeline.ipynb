```python
#number of sample sequenced per time
from matplotlib import pyplot as plt
import numpy as np

time = ["2021*", "Jan 2020", "July 2020", "Aug 2020", "Oct 2020", 
        "Jan 2021", "April 2021", "May 2021", "June 2021", "July 2021"]
number = [12, 1, 2, 4, 1, 5, 6, 37, 38, 21]

def addlabels(time,number):
    for i in range(len(time)):
        plt.text(i,number[i]//0.99,number[i], ha = "center")
        
fig, ax = plt.subplots(figsize = (9,5.5))

addlabels(time, number)
ax.bar(time, number, color = "#c99642", width = 0.6)
ax.set_xlabel("Time")
ax.set_ylabel("Number of sample collected")
ax.set_yticks(np.arange(0,45,5))
ax.margins(x = 0.005)
ax.tick_params(axis = "x", rotation = 60)

plt.tight_layout()
plt.savefig("timeline_sampling.png", dpi = 500)
plt.show()
```
